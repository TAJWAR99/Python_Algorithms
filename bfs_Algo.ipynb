```python
#problem 2
def bfs(graph, start, goal, node):
    visited = []
    queue = []
    dest = {}
    for i in range(node):
        dest[i] = 0
    
    queue.append(start)
    visited.append(start)
    
    while queue:      
        parent = queue.pop(0)
        #print (parent, end = " ") 

        for child in graph[parent]:
            if child not in visited:
                queue.append(child)
                visited.append(child)
                dest[child] = dest[parent] + 1

        #print(dest)
        if goal in visited:
            break
        
    return dest[goal]
    
with open('bfs2.txt') as f:
    node = int(f.readline())
    connection = int(f.readline())
    
    graph = []
    for i in range(node):
        graph.append([])
    
    for i in range(connection):
        u,v = list(map(int,f.readline().split()))
        graph[u].append(v)
        graph[v].append(u)
#     for i in range(node):
#         print(i,"=>",graph[i])
    
    goal = int(f.readline())
    nora = int(f.readline())
    lara = int(f.readline())
    
    a = bfs(graph, nora, goal, node)
    b = bfs(graph, lara, goal, node)
    
    if a>b :
        print('The winner is: Lara')
    else:
        print('The winner is: Nora')
```

```python
#problem 3
def bfs(graph, visited, queue, dest, root, participants):
    queue.append(root)
    visited.append(root)
    stop = False
    
    while queue:      
        parent = queue.pop(0)
        #print (parent, end = " ") 

        for child in graph[parent]: 
            if child not in visited:
                queue.append(child)
                visited.append(child)
                dest[child] = dest[parent] + 1

#         print(dest)
#         print(visited)
        for p in participants:
            if p in visited:
                stop = True
                break
        if stop is True:
            break
        
    print("Minimum number of moves: ",dest[p])
    
with open('bfs3.txt') as f:
    node = int(f.readline())
    connection = int(f.readline())
    
    graph = []
    for i in range(node):
        graph.append([])
    
    for i in range(connection):
        u,v = list(map(int,f.readline().split()))
        graph[v].append(u)
#     for i in range(node):
#         print(i,"=>",graph[i])
    
    goal = int(f.readline())
    number = int(f.readline())
    participants = [int(f.readline()) for i in range(number)]
#     print(participants)
    
    visited = []
    queue = []
    dest = {}
    for i in range(node):
        dest[i] = 0
        
    bfs(graph, visited, queue, dest, goal, participants)
```
