```python
class BinaryNode():
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
class BST:
    def __init__(self):
        self.root = None
    def addNode(self,data):
        if self.root == None:
            node = BinaryNode(data)
            self.root = node
        else:
            self._addNode(self.root,data)
            
    def _addNode(self,current,data):
        if data<current.data:
            if current.left == None:
                node = BinaryNode(data)
                current.left = node
            else:
                self._addNode(current.left,data)
        else:
            if current.right == None:
                node = BinaryNode(data)
                current.right = node
            else:
                self._addNode(current.right,data)
        
    def visit(self,node):
        print(node.data,end=" ")
        
    def in_order(self,current):
        if current != None:
            self.in_order(current.left)
            self.visit(current)
            self.in_order(current.right)
        
    def pre_order(self,current):
        if current != None:
            self.visit(current)
            self.pre_order(current.left)
            self.pre_order(current.right)
        
    def post_order(self,current):
        if current != None:
            self.post_order(current.left)
            self.post_order(current.right)
            self.visit(current)
    
    def printTree(self):
        self.in_order(self.root)
        print(":IN ORDER")
        self.pre_order(self.root)
        print(":PRE ORDER")
        self.post_order(self.root)
        print(":POST ORDER")
    
    def _getHeight(self,node):
        if node == None:
            return 1
        elif node.left and node.right:
            return 1+max(self._getHeight(node.left),self._getHeight(node.right))
        elif node.left:
            return 1+self._getHeight(node.left)
        elif node.right:
            return 1+self._getHeight(node.right)
        else:
            return 1
        
    def getHeight(self):
        if self.root == None:
            print(0)
        else:
            left = self.root.left
            right = self.root.right
            a = self._getHeight(left)
            b = self._getHeight(right)
            if a>=b:
                print(a)
            else:
                print(b)

        
bst = BST()
lst = [3]

for i in lst:
    bst.addNode(i)
bst.printTree()
bst.getHeight()
```

```python
from collections import deque
class Node():
    def __init__(self,value):
        self.value = value
        self.left = self.right = None
class Bst():
    def __init__(self):
        self.root = None
        self.d = deque()
    def insert(self,value):
        node = Node(value)
        if self.root == None:
            self.root = node
        else:
            self._insert(self.root,node)
    def _insert(self,curr,node):
        if node.value < curr.value:
            if curr.left == None:
                curr.left = node
            else:
                self._insert(curr.left,node)
        else:
            if curr.right == None:
                curr.right = node
            else:
                self._insert(curr.right,node)
                
    
    def level_Traverse(self):
        if bool(self.d) == False: #if not d:
            self.d.append(self.root)
        while self.d:
            node = self.d.popleft()
            print(node.value,end=" ")
            if node.left:
                self.d.append(node.left)
            if node.right:
                self.d.append(node.right)
                
t = Bst()
for i in [3,5,4,7,2,1]:
    t.insert(i)
    
t.level_Traverse()
```
