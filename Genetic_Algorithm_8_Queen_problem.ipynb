```python
import numpy as np
import random

def attacking_pair(p):
    lst = []
    for i in range(len(p)):
        attacking_pair = 0
        for j in range(n-1):
            attack = 0
            check = p[i][j]
            c = 1
            for k in range(j+1,n):
                if p[i][k] == check or p[i][k] == check+c or p[i][k] == check-c:
                    attack += 1
                c += 1
            attacking_pair += attack
        lst.append(attacking_pair)
        
    fit_list = fitness(lst)
    return fit_list

def fitness(lst):
    fitness_list = []
    peak_fitness = int(n*(n-1)/2)
    for i in lst:
        fitness_list.append(peak_fitness-i)
    return fitness_list

def probability(fitness_list):
    prob_list = []
    for j in fitness_list:
        prb = (j/sum(fitness_list))
        prob_list.append(prb)
    return prob_list

def selection(prob_list,p):
    arr = [*range(0,len(p))]
    p1,p2 = np.random.choice(arr, 2, True, prob_list)
    return p1,p2

def crossover(p1,p2,arr,p):
    x,y = p[p1],p[p2]
    num = random.randint(0,len(p)-1)
    
    child_1 = np.append(x[:num],y[num:],axis=0)
    child_2 = np.append(y[:num],x[num:],axis=0)
    arr.append(child_1)
    arr.append(child_2)
    return arr

def mutation(arr,mutation_threshold):
    for i in arr:
        if np.random.uniform(0, 1) < mutation_threshold:
            index = random.randint(0,7)
            new_num = random.randint(0,7)
            i[index] = new_num
    return arr
        
def GA(p, n, mutation_threshold):
        ans = -1
        fit = attacking_pair(p)
        for indx,val in enumerate(fit):
            if val == int(n*(n-1)/2):
                ans = indx
                return p,ans
                        
        prob_list = probability(fit)

        arr = []
        for i in range(len(p)//2):
            p1,p2 = selection(prob_list,p)

            arr = crossover(p1,p2,arr,p)
        arr = np.array(arr)
        arr = mutation(arr,mutation_threshold)
        
        #adding old generations with new ones
        
        p1,p2 = np.random.choice(len(p),2)
        new_arr = []
        new_arr.append(p[p1])
        new_arr.append(p[p2])
        new_arr = np.array(new_arr)
        arr = np.append(arr,new_arr,axis=0)
        return arr,ans


#--------------------------------------------
n = 8
start_population = 10
mutation_threshold = 0.3

p = np.random.randint(0, n, (start_population, n))
# print(p)

while True:
    p,ans =  GA(p, n, mutation_threshold)
    if ans != -1:
        break
        
print(p[ans])
```
